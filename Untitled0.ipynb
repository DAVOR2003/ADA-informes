**1.Problema de Ordenamiento**

Un sistema, sin importar su tamaño, funcionará mejor cuando todo está en perfecto orden. Es por esto que se requiere que los datos se encuentren en un orden óptimo para ser utilizados por lo que en este informe se expondrá una de las soluciones ya existentes para el ordenamiento de datos, el cual recibe el nombre de **"Insertion Sort"**.

A continuación, se definen los datos de entrada y salida del algoritmo que solucionará el problema.

Entrada: Arreglo o Lista de N datos comparables.

*Si se ingresa un arreglo o lista de datos heterogeneos, el algoritmo  no será  capaz de ordenarlos.*

Salida: Arreglo o Lista con los N datos ordenados de forma ascendente.

**2. Descripción del Insertion Sort**
El algortimo recibe como parametro un arreglo o lista con n elementos comparables los cuales se ordenarán dentro de la misma estructura.


1. Por cada iteración i, el algoritmo recorre el arreglo comparando la posición i+1 con la posición que le precede, intercambiándose los datos de tal forma que los datos estén ordenados de menor mayor al finalizar cada iteración.

2. Si el arreglo se encuentra ordenado, el algoritmo termina. Si no, se vuelve a iterar lo explicado en el paso 1.

3. Al finalizar el ordenamiento, se retorna el arreglo o lista ya ordenado de menor a mayor.

Cabe destacar que la cantidad de comparaciones, dependerá solamente del dato que se quiere ordenar, ya que si es un dato menor a la gran parte de los ya ordenados, se realizarán una gran cantidad de comparaciones para encontrar el sitio correcto de este. Análogamente, se realizarán una menor cantidad de comparaciones si el dato es mayor a la media de datos ya ordenados.
